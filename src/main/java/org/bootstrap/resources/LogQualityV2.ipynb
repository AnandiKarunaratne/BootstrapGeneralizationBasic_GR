```python
import csv
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
# calculate precision and recall for noise 0%

precision_means_0 = []
precision_confidences_upper_0 = []
precision_confidences_lower_0 = []
recall_means_0 = []
recall_confidences_upper_0 = []
recall_confidences_lower_0 = []
for i in [100, 1000, 10000, 100000]:
    precision_results = []
    recall_results = []
    with open('/Users/anandik/Documents/GR/Code/BootstrapGeneralizationBasic_GR/src/main/java/org/bootstrap/resources/omitactivity/output-' + str(i) +'-0.csv', 'r') as csvfile:
        csvreader = csv.reader(csvfile)
        
        for row in csvreader:
            if len(row) >= 1:
                precision_value = float(row[0])
                recall_value = float(row[1])

                precision_results.append(precision_value)
                recall_results.append(recall_value)

    precision_mean = np.mean(precision_results)
    precision_sem = stats.sem(precision_results)
    recall_mean = np.mean(recall_results)
    recall_sem = stats.sem(recall_results)
    precision_confidence_interval = stats.t.interval(0.95, len(precision_results) - 1, loc=precision_mean, scale=precision_sem)
    recall_confidence_interval = stats.t.interval(0.95, len(recall_results) - 1, loc=recall_mean, scale=recall_sem)
    precision_means_0.append(precision_mean)
    precision_confidences_lower_0.append(precision_confidence_interval[0])
    precision_confidences_upper_0.append(precision_confidence_interval[1])
    recall_means_0.append(recall_mean)
    recall_confidences_lower_0.append(recall_confidence_interval[0])
    recall_confidences_upper_0.append(recall_confidence_interval[1])
    
print(precision_means_0)
print(recall_means_0)
print(precision_confidences_upper_0)
print(precision_confidences_lower_0)
print(recall_confidences_upper_0)
print(recall_confidences_lower_0)
```
